# Facial Emotion Recognition

### Import required modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import os
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
from keras.models import Sequential
from keras.layers import Dense,Dropout,Flatten,MaxPooling2D,Conv2D
from tqdm.notebook import tqdm
from keras.preprocessing.image import load_img
from keras.utils import to_categorical
```

### Load the Dataset

```python
TRAIN_DIR='../Facial Emotion Recognition/train/train/'
TEST_DIR='../Facial Emotion Recognition/test/test/'
```

```python
def load_dataset(directory):
    image_paths=[]
    labels=[]
    for label in os.listdir(directory):
        for filename in os.listdir(directory+label):
            image_path=os.path.join(directory,label,filename)
            image_paths.append(image_path)
            labels.append(label)
        print(label,"completed")
    return image_paths,labels
            
```

```python
# convert into dataframe - train
train=pd.DataFrame()
train['image'],train['label']=load_dataset(TRAIN_DIR)

# shuffle the dataset - test
train=train.sample(frac=1).reset_index(drop=True)
```

```python
# convert into dataframe
test=pd.DataFrame()
test['image'],test['label']=load_dataset(TEST_DIR)
```

### Feature Extraction

```python
from keras.preprocessing.image import img_to_array, array_to_img
from tqdm.notebook import tqdm

def resize_images(images, target_size=(48, 48)):
    resized_images = []
    for image in tqdm(images):
        img = load_img(image)
        img = img.resize(target_size)
        img_array = img_to_array(img)
        
        # Check if the image is correctly resized
        if img_array.shape != (48, 48, 3):  # 3 channels for RGB images
            print(f"Image {image} is not correctly resized, it has shape {img_array.shape}")
        
        resized_images.append(img_array)
    
    resized_images = np.array(resized_images)
    
    # Check if all images have the correct shape
    for img in resized_images:
        if img.shape != (48, 48, 3):
            print(f"Found an image with incorrect shape: {img.shape}")
    
    return resized_images

train_features = resize_images(train['image'])

print(f"Shape of train_features before reshaping: {train_features.shape}")
train_features = np.mean(train_features, axis=-1, keepdims=True)
print(f"Shape of train_features after converting to grayscale: {train_features.shape}")
train_features = train_features.reshape(len(train_features), 48, 48, 1)  
print(f"Train features shape: {train_features.shape}")

```

```python
test_features = resize_images(test['image'])
print(f"Shape of test_features before reshaping: {test_features.shape}")
test_features = np.mean(test_features, axis=-1, keepdims=True)
print(f"Shape of test_features after converting to grayscale: {test_features.shape}")
test_features = test_features.reshape(len(test_features), 48, 48, 1)  
print(f"Test features shape: {test_features.shape}")
```

```python
# normalise the image
X_train=train_features/255.0
X_test=test_features/255.0
```

```python
# convert label names to integer
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(train['label'])
Y_train=le.transform(train['label'])
Y_test=le.transform(test['label'])
```

```python
Y_train=to_categorical(Y_train,num_classes=7)
Y_test=to_categorical(Y_test,num_classes=7)
```

```python
# config
input_shape=(48,48,1)
output_class=7
```

### Model Creation

```python
model=Sequential()
model.add(Conv2D(128,kernel_size=(3,3),activation='relu',input_shape=input_shape))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))
          
model.add(Conv2D(256,kernel_size=(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))
          
model.add(Conv2D(512,kernel_size=(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))
          
model.add(Conv2D(512,kernel_size=(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))

model.add(Flatten())
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(256,activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(output_class,activation='softmax'))

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])                                                                 
```

```python
history=model.fit(x=X_train,y=Y_train,batch_size=128,epochs=2,validation_data=(X_test,Y_test))
```
